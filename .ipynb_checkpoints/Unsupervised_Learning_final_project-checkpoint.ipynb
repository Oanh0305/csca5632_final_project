## Data cleaning

* Removing stopwords
* Filter out short words
* Lowercase and removing everything except words
* Applying stemming vs lemmatization to the text

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
stopWords = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

stopWords = set(stopwords.words('english'))

def textcleaner_lemmas(text):
    ''' Takes in raw unformatted text and strips punctuation, removes whitespace,
    strips numbers, tokenizes and stems.
    Returns string of processed text to be used into CountVectorizer
    '''
    # Lowercase and strip everything except words
    cleaner = re.sub(r"[^a-zA-Z ]+", ' ', text.lower())
    # Tokenize
    cleaner = word_tokenize(cleaner)
    ps = PorterStemmer()
    clean = []
    for w in cleaner:
        # filter out stopwords
        if w not in stopWords:
            # filter out short words
            if len(w)>2:
                # lemmatizer 
                clean.append(lemmatizer.lemmatize(w))
    return ' '.join(clean)
```

```python
df_full['clean_text_lemma'] = df_full.text.apply(lambda x: textcleaner_lemmas(x))
df_train['clean_text_lemma'] = df_train.text.apply(lambda x: textcleaner_lemmas(x))
df_test['clean_text_lemma'] = df_test.text.apply(lambda x: textcleaner_lemmas(x))
```

### Topic important words based on NMF probabilities

```python
# Number of topics.

n_topics=20

# Linking words to topics
def word_topic(tfidf, topic_probs, wordlist):
    
    # Loading scores for each word on each topic/component.
    words_by_topic = tfidf.T * topic_probs

    # Linking the loadings to the words in an easy-to-read way.
    components = pd.DataFrame(words_by_topic, index = wordlist)

    return components
    
# Extracts the top N words and their loadings for each topic.
def top_words(components, n_top_words):
    n_topics = range(components.shape[1])
    index= np.repeat(n_topics, n_top_words, axis=0)
    top_words=pd.Series(index=index)
    for column in range(components.shape[1]):
        sortedwords = components.iloc[:,column].sort_values(ascending=False)
        chosen = sortedwords[:n_top_words]
        chosen_list = chosen.index +" "+round(chosen,2).map(str) 
        top_words.loc[column]=[x for x in chosen_list]
    return(top_words)
```

```python
vectorizer = TfidfVectorizer(min_df=6, strip_accents='ascii', analyzer='word', lowercase=True,
                             ngram_range=(1,2))

x_train_lemma = vectorizer.fit_transform(df_train['clean_text_lemma'])
y_train_lemma = df_train['source']
x_test_lemma = vectorizer.transform(df_test['clean_text_lemma'])
y_test_lemma = df_test['source']
features_train = vectorizer.get_feature_names_out()

terms = vectorizer.get_feature_names_out()
```

## Non-Negative Matrix Factorization Models

### Model NMF using Frobenius Loss

```python
from sklearn.decomposition import NMF

nmf = NMF(alpha_W=0.0, 
          init='nndsvdar', # how starting value are calculated
          l1_ratio=0.0, # Sets whether regularization is L2 (0), L1 (1), or a combination (values between 0 and 1)
          max_iter=200, # when to stop even if the model is not converging (to prevent running forever)
          n_components=20,
          beta_loss='frobenius', 
          random_state=0, 
          solver='cd', # Use Coordinate Descent to solve
          tol=0.0001, # model will stop if tfidf-WH <= tol
          verbose=0 # amount of output to give while iterating
         )
news_group_nmf = nmf.fit_transform(x_train_lemma) 

components_nmf = word_topic(x_train_lemma, news_group_nmf, terms)

topwords = pd.DataFrame()

topwords['NNMF']=top_words(components_nmf, 10)
```

### Grouping labels for a simpler classification

Since we are having 20 news group and lots of them are similar to each other, running clustering on these data are very difficult. Therefore, in the next step, we will group the labels to label groups that are related to each other.

We decide the labels into 7 different categories:
- Politics: talk.politics.mideast, talk.politics.guns, talk.politics.misc
- Sport: rec.sport.baseball, rec.sport.hockey
- Computer: comp.graphics, comp.windows.x, comp.sys.mac.hardware, comp.os.ms-windows.misc, comp.sys.ibm.pc.hardware
- Science: sci.crypt, sci.med, sci.space, sci.electronics
- Religions: alt.atheism, soc.religion.christian, talk.religion.misc
- Sales: misc.forsale, 
- Autos: rec.motorcycles, rec.autos

```python
group_to_labels = {'Politics': ['talk.politics.mideast', 'talk.politics.guns', 'talk.politics.misc'],
                  'Sport': ['rec.sport.baseball', 'rec.sport.hockey'],
                  'Computer': ['comp.graphics', 'comp.windows.x', 'comp.sys.mac.hardware', 'comp.os.ms-windows.misc', 'comp.sys.ibm.pc.hardware'],
                  'Science': ['sci.crypt', 'sci.med', 'sci.space', 'sci.electronics'],
                  'Religions': ['alt.atheism', 'soc.religion.christian', 'talk.religion.misc'],
                  'Sales': ['misc.forsale'],
                  'Autos': ['rec.motorcycles', 'rec.autos']}

label_to_group = {}
for group_name, group in group_to_labels.items():
    for label in group:
        label_to_group[label] = group_name
        
df_train['new_label'] = df_train['label'].apply(lambda x: label_to_group[x])
df_test['new_label'] = df_test['label'].apply(lambda x: label_to_group[x])
```

```python
df_train['new_label'] 
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_train['new_label_num'] = label_encoder.fit_transform(df_train['new_label'])
df_test['new_label_num'] = label_encoder.transform(df_test['new_label'])
```

```python
from sklearn.decomposition import NMF

nmf = NMF(n_components=5, 
    random_state=42,
    init='nndsvda',
    beta_loss='frobenius',
    max_iter=200,
    alpha_W=0.00005,
    alpha_H=0.00005,
    l1_ratio=0)

news_group_nmf = nmf.fit_transform(x_train_lemma) 
train_pred = news_group_nmf.argmax(axis=1)
```

```python
# Since the label output by nmf is unordered, we will find the best 
# label permutation that yields the highest accuracy

import itertools
def label_permutation_acc(yt, yp, n): 
    # intialize several parameters
    best_accuracy=float('-inf')
    best_permutation=()
    best_label_mapping_yp = {}
    best_ypred=[]
    permutations=list(itertools.permutations(np.arange(n))) # a list of size n tuples 

    # get actual 'Category' ['business', 'tech', 'politics', 'sport', 'entertainment']
    unique_labels= yt.unique() 

    # map 'numerical labels' to 'Category': Ex:  {0: 'business', '1: 'tech', 2: 'politics', 3: 'sport', : 'entertainment'}
    label_mapping_yt = {i: label  for i,label in enumerate(unique_labels)}

    for permutation in permutations:
        label_mapping_yp = {perm:  label_mapping_yt[i]  for  i,perm in enumerate(list(permutation))}
        yp_gt = [label_mapping_yp[y] for y in yp] # transfer to ground truth label
        accuracy = (yt==yp_gt).sum()/len(yt) 
        # update to keep the best accuracy permutation
        if accuracy> best_accuracy:
            best_permutation = permutation
            best_accuracy=accuracy
            best_ypred = yp_gt
            best_label_mapping_yp = label_mapping_yp

    return (best_label_mapping_yp, best_accuracy)

best_mapping, acc = label_permutation_acc(df_train['new_label'], train_pred, n=6)
print('Training accuracy score: ', acc)
```

```python
test_news_group_nmf = nmf.fit_transform(x_test_lemma) 
test_pred = test_news_group_nmf.argmax(axis=1)

accuracy_score(df_test['new_label'], [best_mapping[x] for x in test_pred])
```

```python
nmf_2 = NMF(alpha_W=0.0, 
          init='nndsvdar', # how starting value are calculated
          l1_ratio=0.0, # Sets whether regularization is L2 (0), L1 (1), or a combination (values between 0 and 1)
          max_iter=200, # when to stop even if the model is not converging (to prevent running forever)
          n_components=5,
            solver ='mu',
          beta_loss='kullback-leibler', 
          random_state=0, 
          tol=0.0001, # model will stop if tfidf-WH <= tol
          verbose=0 # amount of output to give while iterating
         )
```

```python
train_nmf_2 = nmf_2.fit_transform(x_train_lemma) 

train_pred_2 = train_nmf_2.argmax(axis=1)

best_mapping, acc = label_permutation_acc(df_train['new_label'], train_pred_2, n=6)
print('Training accuracy score: ', acc)
```

```python
test_news_group_nmf = nmf.fit_transform(x_test_lemma) 
test_pred = test_news_group_nmf.argmax(axis=1)

accuracy_score(df_test['new_label'], [best_mapping[x] for x in test_pred])
```

## Conclusion


Because our problem is very hard since there are too many labels (20 labels in total) so it is very hard for unsuppervised learning like NMF to have good performance. First, I used model NMF using Frobenius loss and Kullback-Leibler Divergence which gave us very low performance( 0.06 and 0.08) on test dataset.
After that I decided to make grouping labels for a simpler classification, accuracy score we got higher to 0.36. This result is still not good but that the best I can do here. Besides, I also ran supervised method like random forest classifier which is much higher 0.56 on test dataset.We can see that, since supervised method has labels when training so it gave higher performance when compare to unsupervised method which has no labels when training.
